```python
import cv2
import urllib.request
from google.colab.patches import cv2_imshow
from ultralytics import YOLO
from collections import Counter
import numpy as np
import time
import math

# 이미지 URL
url = 'https://img.etnews.com/photonews/2007/1317575_20200707171810_433_0002.jpg'

# 이미지 다운로드
urllib.request.urlretrieve(url, "road_image.jpg")

# 모델 로드
model = YOLO('yolov8n.pt')

# 객체 카운터 및 속도 추정을 위한 딕셔너리 초기화
object_counter = Counter()
vehicle_positions = {}
vehicle_speeds = {}
vehicle_directions = {}

# 픽셀 당 미터 (예시 값, 실제 환경에 맞게 조정 필요)
PIXELS_PER_METER = 0.1

# 프레임 레이트 (예시 값, 실제 환경에 맞게 조정 필요)
FPS = 30

# 방향 결정을 위한 각도 범위
ANGLE_RANGES = {
    'North': (315, 45),
    'East': (45, 135),
    'South': (135, 225),
    'West': (225, 315)
}

def get_direction(angle):
    for direction, (start, end) in ANGLE_RANGES.items():
        if start <= angle < end:
            return direction
    return 'North'  # 기본값

# 이미지에서 객체 감지
results = model('road_image.jpg')

# 이미지 읽기
img = cv2.imread('road_image.jpg')

# 색상 정의
color_dict = {
    'car': (0, 255, 0),    # 초록색
    'bus': (0, 0, 255),    # 빨간색
    'truck': (0, 255, 255) # 노란색
}

# 결과 시각화 및 속도, 방향 추정
for r in results:
    boxes = r.boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        cls = int(box.cls)
        class_name = model.names[cls]

        # 색상 선택 (기본값은 흰색)
        color = color_dict.get(class_name, (255, 255, 255))

        # 객체 중심점 계산
        center_x = (x1 + x2) / 2
        center_y = (y1 + y2) / 2

        # 객체 ID (예시로 중심점 좌표를 ID로 사용)
        object_id = f"{class_name}_{int(center_x)}_{int(center_y)}"

        # 이전 위치가 있으면 속도와 방향 계산
        if object_id in vehicle_positions:
            prev_x, prev_y, prev_time = vehicle_positions[object_id]
            dx = center_x - prev_x
            dy = center_y - prev_y
            distance = np.sqrt(dx**2 + dy**2) * PIXELS_PER_METER
            time_diff = time.time() - prev_time
            speed = distance / time_diff * 3.6  # m/s를 km/h로 변환
            vehicle_speeds[object_id] = speed

            # 방향 계산
            angle = math.degrees(math.atan2(-dy, dx))  # 양의 y축이 아래로 향하므로 -dy 사용
            angle = (angle + 360) % 360  # 0-360 범위로 조정
            direction = get_direction(angle)
            vehicle_directions[object_id] = direction

        # 현재 위치 및 시간 저장
        vehicle_positions[object_id] = (center_x, center_y, time.time())

        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
        label = f'{class_name}'
        if object_id in vehicle_speeds:
            label += f' {vehicle_speeds[object_id]:.1f} km/h'
        if object_id in vehicle_directions:
            label += f' {vehicle_directions[object_id]}'
        cv2.putText(img, label, (int(x1), int(y1) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

        # 객체 카운트
        object_counter[class_name] += 1

# 결과 표시
cv2_imshow(img)

# 객체 수 출력
print("검출된 객체:")
for obj, count in object_counter.items():
    print(f"{obj}: {count}대")
```
