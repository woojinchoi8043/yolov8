```python
import cv2
from ultralytics import YOLO
from collections import Counter
from google.colab.patches import cv2_imshow
import numpy as np

# YOLO 모델 로드
model = YOLO('yolov8x.pt')

# 비디오 파일 경로 (로컬 비디오 파일 경로를 입력하세요)
video_path = 'https://www.bing.com/videos/riverview/relatedvideo?&q=%eb%8f%84%eb%a1%9c+cctv&&mid=199AA73118933CDD0A7A199AA73118933CDD0A7A&&FORM=VRDGAR'

# 비디오 캡처 객체 생성
cap = cv2.VideoCapture(video_path)

# 색상 정의
color_dict = {
    'car': (0, 255, 0),    # 초록색
    'bus': (0, 0, 255),    # 빨간색
    'truck': (0, 255, 255) # 노란색
}

# 비디오 저장 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('/content/output.mp4', fourcc, fps, (width, height))

processed_frames = []

while cap.isOpened():
    # 프레임 읽기
    success, frame = cap.read()
    if not success:
        break

    # 객체 감지
    results = model(frame)

    # 객체 카운터 초기화
    object_counter = Counter()

    # 결과 시각화
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
            cls = int(box.cls)
            class_name = model.names[cls]

            # 차량 종류만 처리 (car, bus, truck)
            if class_name in ['car', 'bus', 'truck']:
                # 색상 선택
                color = color_dict.get(class_name, (255, 255, 255))

                # 바운딩 박스 그리기
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)
                cv2.putText(frame, f'{class_name}', (int(x1), int(y1) - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

                # 객체 카운트
                object_counter[class_name] += 1

    # 객체 수 표시
    y_offset = 30
    for obj, count in object_counter.items():
        cv2.putText(frame, f'{obj}: {count}', (10, y_offset),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        y_offset += 30

    # 프레임 저장
    out.write(frame)

    # 프레임 수집 (선택적)
    processed_frames.append(frame)

# 자원 해제
cap.release()
out.release()

# 처리된 비디오 미리보기 (선택적)
for frame in processed_frames[:10]:  # 처음 10프레임만 표시
    cv2_imshow(frame)

print("검출 영상이 '/content/output.mp4'로 저장되었습니다.")
```
